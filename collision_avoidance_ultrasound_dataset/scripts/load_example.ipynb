```python
idx_collection = 0  # collection to load
verbose = True  # desired verbosity

x, y = usld.load_collection(
    idx_collection=idx_collection,
    verbose=verbose,
)
```

```python
# The y NumPy array contains all the labels of the collection's windows.

print(y.shape)
print(y.dtype)
print(y)
```

```python
# The x NumPy array contains all the collection's signal windows.
# Each window has 9 channels, 2048 samples, and uint8 values.

print(x.shape)
print(x.dtype)

idx_window_to_plot = 0
idx_channel_to_plot = 0
x_to_plot = x[idx_window_to_plot, idx_channel_to_plot]

NUM_SAMPLES = 2048  # samples of each window
FS_HZ = 100e3  # dataset's sampling frequency: 100kHz
times_s = np.arange(NUM_SAMPLES, dtype=np.float32) / FS_HZ  # time values in s
time_max_s = NUM_SAMPLES / FS_HZ

plt.figure(figsize=(10.0, 4.0))

plt.title(
    f"COLLECTION {idx_collection:d} - Time window {idx_window_to_plot:d}\n"
    f"Channel {idx_channel_to_plot:d} (zero-based numbering)"
)

plt.xlabel("Time (s)")
plt.ylabel("Ultrasound signal\n(uint8 ADC values)")
plt.plot(times_s, x_to_plot, 'b-')
plt.grid()
plt.axis([0.0, time_max_s, 0.0, 256.0])
plt.show()
```
